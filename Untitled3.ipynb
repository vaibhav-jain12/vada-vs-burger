```python
import os
import shutil
import random
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers,models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
from google.colab import drive
drive.mount('/content/drive')

SOURCE_DIR = "/content/drive/MyDrive/VadaPav"
WORK_DIR = "/content/data"

TRAIN_DIR = os.path.join(WORK_DIR, "train")
VAL_DIR   = os.path.join(WORK_DIR, "val")

CLASSES = {
    "burger": "burger",
    "vada pav": "vada_pav"
}


for split in [TRAIN_DIR, VAL_DIR]:
    for cls in CLASSES.values():
        os.makedirs(os.path.join(split, cls), exist_ok=True)


SPLIT_RATIO = 0.8
random.seed(42)

for src_cls, dst_cls in CLASSES.items():
    src_path = os.path.join(SOURCE_DIR, src_cls)
    images = os.listdir(src_path)
    random.shuffle(images)

    split_idx = int(len(images) * SPLIT_RATIO)

    for i, img in enumerate(images):
        src_img = os.path.join(src_path, img)

        if i < split_idx:
            dst_img = os.path.join(TRAIN_DIR, dst_cls, img)
        else:
            dst_img = os.path.join(VAL_DIR, dst_cls, img)

        shutil.copy(src_img, dst_img)

print("✅ Dataset split completed")

```

```python
IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    brightness_range=[0.8, 1.2]
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    "/content/data/train",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary"
)

val_generator = val_datagen.flow_from_directory(
    "/content/data/val",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary"
)
```

```python
model = models.Sequential([

    layers.Conv2D(32,(3,3), padding="same" , input_shape=(224,224,3)),
    layers.ReLU(),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64,(3,3), padding="same"),
    layers.ReLU(),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128,(3,3), padding="same"),
    layers.ReLU(),
    layers.MaxPooling2D(2,2),

    layers.GlobalAveragePooling2D(),
    layers.Dense(128),
    layers.ReLU(),
    layers.Dropout(0.5),
    layers.Dense(1, activation="sigmoid")
])
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
from PIL import Image
import os

def check_images_for_corruption(directory):
    corrupted_files = []
    print(f"Checking images in: {directory}")
    for root, _, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
                filepath = os.path.join(root, file)
                try:
                    with Image.open(filepath) as img:
                        img.load() # Attempt to fully load the image data for a more thorough check
                except (IOError, SyntaxError) as e:
                    print(f"Corrupted image detected: {filepath} - {e}")
                    corrupted_files.append(filepath)
                except Exception as e:
                    print(f"Unexpected error loading image: {filepath} - {e}")
                    corrupted_files.append(filepath)
    return corrupted_files

# Check both training and validation directories
corrupted_train_images = check_images_for_corruption(TRAIN_DIR)
corrupted_val_images = check_images_for_corruption(VAL_DIR)

if corrupted_train_images or corrupted_val_images:
    print("\nFound corrupted images. Please remove or replace them:")
    for f in corrupted_train_images:
        print(f)
    for f in corrupted_val_images:
        print(f)
    print("After addressing these files, please re-run the dataset splitting cell (cell NE91x7v59_O5) and then the model training cell (cell oxMPZN9GFrrJ).")
else:
    print("No corrupted images found. You can proceed with model training.")
```

```python
from google.colab import files
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import io

def test_your_model():
    print("Click 'Choose Files' to upload a Burger or Vada Pav image:")
    uploaded = files.upload()

    for filename in uploaded.keys():
        # 1. Load the image from the uploaded bytes
        img = image.load_img(io.BytesIO(uploaded[filename]), target_size=(224, 224))

        # 2. Preprocess
        img_array = image.img_to_array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        # 3. Predict
        prediction = model.predict(img_array)

        # 4. Display result
        plt.figure(figsize=(4, 4))
        plt.imshow(img)
        plt.axis('off')

        # Classes: 0 is likely Burger, 1 is Vada Pav (based on ImageDataGenerator's alphabetical order)
        # train_generator.class_indices usually maps {'burger': 0, 'vada_pav': 1}
        if prediction[0] > 0.5:
            result = f"Vada Pav (Confidence: {prediction[0][0]*100:.1f}%)"
        else:
            result = f"Burger (Confidence: {(1-prediction[0][0])*100:.1f}%)"

        plt.title(result)
        plt.show()

# Run the function
test_your_model()
```
